```python
data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
```

```python
test = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
```

```python
data
```

```python
y = data['label']
data = data.drop(columns = 'label', axis =1)
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(data),columns = data.columns) 
```

```python
df_test = pd.DataFrame(scaler.fit_transform(test),columns = test.columns) 
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
y_train_one_hot = tf.keras.utils.to_categorical(y_train, num_classes=10)
```

```python
y_train_one_hot
```

```python
y_train
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense , Flatten
from tensorflow.keras.optimizers import Adam
model = Sequential([
    Dense(256, activation='relu', input_shape = (784,)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
optimizer=Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train_one_hot, epochs=20)
```

```python
predictions = model.predict(X_test)
predictions1 = np.argmax(predictions, axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions1)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
preds = model.predict(test)
preds = np.argmax(preds,axis=1)
preds
```

```python
df = pd.DataFrame({'ImageId': test.index + 1 , 'Label': preds})
```

```python
df.to_csv('submission.csv', index=False)
```
